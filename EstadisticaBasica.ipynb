Determina si existe una correlación entre el índice de masa corporal (bmi) y el costo del seguro

```python
from scipy import stats 
from scipy.stats import pearsonr
```

```python
selected = df[['age','bmi','children','charges']]
```

```python
r, p = stats.pearsonr(selected['bmi'], selected['charges'])
print(f"Correlación Pearson: r={r}, p-value={p}")
```

```python
if (p < 0.05):
    print('La hipótesis alterna se confirma')

else:
    print('La hipótesis nula se confirma')
```

En promedio, ¿quién paga más de cuota de seguro? ¿Los fumadores o los no fumadores? Muéstralo con los datos

¿Cuáles son las cuotas mínimas y máximas que las personas pagan dependiendo del género y del número de hijos?
